```python
import numpy as np
import matplotlib.pyplot as plt
```

<font size = 4, color = 'red'>1. Basic Plot Example</font>

This example shows a simple plot of Cosine function. In this plot, I am basically learning to deal with following things:

1) How to make a plot
2) Adding a legend
3) Setting X and Y tick labels and markers
4) Making annotation

```python
x_val = np.linspace(-360, 360, num = 100)
conv_factor = np.pi/180
y = np.cos(x_val*conv_factor)

# Assigning x and y tick labels and tick values #
#----------------------------------------------#

tick_values = np.linspace(-360, 360, 9)
tick_labels = [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
y_tick_values = np.linspace(-1, 1, 9)
y_tick_labels = [-1.0, -0.75, -0.50, -0.25, 0.0, 0.25, 0.50, 0.75, 1.0 ]

fig = plt.figure(figsize = (15, 12), dpi = 100)
ax = fig.add_subplot(1,1,1)

# Setting x and y tick labels and tick values #
#----------------------------------------------#

ax.set_yticks(y_tick_values)
ax.set_xticks(tick_values)
ax.set_xticklabels(tick_labels, size = 15)
ax.set_yticklabels(y_tick_labels, size =15)
ax.plot(x_val,y, lw = 3, color = 'blue')

# Setting x and y labels and title #
#----------------------------------------------#

ax.set_xlim(-370, 370)
ax.set_xlabel('Time', size = 20)
ax.set_ylabel('Distance', size = 20)
ax.set_title('Sinusoidal Motion', size = 20)

# Marking a point using scatter plot #
#----------------------------------------------#

ax.scatter(60, np.cos(60*conv_factor), color = 'red', s = 80)

# Setting legend and annotation of the point #
#----------------------------------------------#

ax.legend(['Cosine Curve'], loc = (0.65, 0.85), frameon = True, fontsize = 'x-large')
ax.annotate(r'$Cos(60) = 0.5$', xy = (60, np.cos(60*conv_factor)), size = 20, textcoords = 'offset points', xycoords = 'data', xytext = (30,-20), arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3, rad = 0.85'))

ax.grid()
plt.show()
plt.close()
```

<font size = 4, color = 'red'>2. Multiple plots in different plot frame</font>

In this section I am trying to show how to make multiple plots using 'subplots' .  All other features are already covered in previous plot example. 

```python
x = np.linspace(-360, 360, num =  100)
conv_factor = np.pi/180
cos = np.cos(x*conv_factor)
tick_values  = np.linspace(-360,360, num = 9)

tick_labels = [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']

sin = np.sin(x*conv_factor)
fig = plt.figure(figsize=(12, 10))

#-----------------------------------------------------------------------------------#
# To make a multiple plot we use subplots() and the number inside bracket indicate number of rows #
# and number of columns. For instance, we have two rows and one column in the plot below. #

ax1, ax2 = fig.subplots(2,1)
cosine_plot = ax1.plot(x,cos)
sine_plot = ax2.plot(x,sin, color = 'red')
ax1.legend(['Cosine Curve'], loc = (0.60,0.75), fontsize = 15)
ax1.set_xlabel('X Values')
ax1.set_ylabel('Y Values')
ax1.set_xticks(tick_values)
ax1.set_xticklabels(tick_labels)
ax1.set_title('Sinusoidal Motion')
ax1.grid()

ax2.set_xticks(tick_values)
ax2.set_xticklabels(tick_labels)
ax2.set_xlabel('X Values')
ax2.set_ylabel('Y Values')
ax2.legend(['Sine Curve'], loc = (0.25,0.75), fontsize = 15)
ax2.grid()
plt.show()
plt.close()
```

<font size = 4, color = 'red'>3. Scatter plots and tilting tick marks </font>

In this plot, I introduce scatter plot and how we can tilt tick marks. 

Note: 

<font color = 'blue'> Scatter plot below is not the best way to visualize the data given. Line graph would be better. Data visualization requires great skills and it is a kind of art form. Only through a properly visualized data we can understand the information data is trying to convey.  </font>

```python
ktm_rainfall = [1621, 1871, 1740, 1583, 1236, 1391, 1346, 1220, 1205, 1479, 1655, 1465]

pkr_rainfall = [4512, 4484, 4362, 4129, 2967, 3024, 4272, 3263, 3256, 3967, 3487, 3266]

year = np.linspace(2001, 2012, num = 12)
tick_values = year


fig = plt.figure(figsize = (15, 10))
ax = fig.subplots()
ax.scatter(year, ktm_rainfall, color = 'red')
ax.scatter(year, pkr_rainfall, color = 'blue') 

ax.set_xlim(2000, 2013)
ax.set_xticks(tick_values)

#--------------------------------#
# Here we learn how to tilt the tick marks #
x_labels = ax.get_xticklabels()
plt.setp(x_labels, rotation = 50)


ax.set_xlabel('Year', size = 15)
ax.set_ylabel('Rainfall (in mm)',  size = 15)
ax.set_title('Rainfall in Kathmandu and Pokhara from year 2001 to 2012', size = 20)
ax.legend(['Kathmandu','Pokhara'], loc = (0.03, 0.5), fontsize = 20)



plt.show()
plt.close()
```
